# **Image Processing**


**Shift**

```
#pic=cv2.imread('Planet.png',1) #with color 
#pic=cv2.imread('Planet.png',0) #with no color
pic=cv2.imread('Planet.png',-1) # with alpha 
cols = pic.shape[1]
rows = pic.shape[0]

M= np.float32([[-1,0,350],[0,1,200]]) # Translation Matrix

shifted= cv2.warpAffine(pic,M, (cols,rows))
cv2_imshow(shifted)
```

**Thresholding**

```
pic1=cv2.imread('Planet.png',0)
threshold_value=150
(T_value,binary_threshold) =cv2.threshold(pic1,threshold_value,255,cv2.THRESH_BINARY)
cv2_imshow(binary_threshold)
```

**Median Blur**

```
pic3 = cv2.imread('face.jpg',0)

median_blur=cv2.medianBlur(pic3,9)
cv2_imshow(median_blur)
```

```
pic4 = cv2.imread('face.jpg',1)

dim_pixel=9
color=150
space=150

bilateral_filter=cv2.bilateralFilter(pic4,dim_pixel,color,space)
cv2_imshow(bilateral_filter)
```
